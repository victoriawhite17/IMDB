```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Data

```python
#loading data for movies started in 2000
first_year_df = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
first_year_df.head()
```

```python
#loading data for movies started in 2001
second_year_df = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
second_year_df.head()
```

```python
final_year_df = pd.concat([first_year_df, second_year_df])
final_year_df.head()
```

```python
final_year_df.info()
```

###  Revenue OR Budget > 0

```python
#creating filter for budget > 0 and revenue >0
budget_filter = final_year_df['budget'] > 0
revenue_filter = final_year_df['revenue'] > 0
#filtering movie with budget OR revenue > 0
revenue_or_budget = final_year_df.loc[budget_filter | revenue_filter, :]
revenue_or_budget
```

####  There are 681 movies that had a budget or revenue that was more than $0.

### Movies with Ratings

```python
#filtering movies for budget AND revenue > 0
filtered_df = final_year_df.loc[budget_filter & revenue_filter, :]
filtered_df
```

```python
filtered_df['certification'].value_counts()
```

####  There are 2 movies rated G, 8 movies rated PG, 10 movies rated PG-13, and 14 movies rated R.

### Finding Average Revenue by Movie Rating

```python
#creating filters for movie ratings
rated_r_filter = filtered_df['certification'] == 'R'
rated_pg_13_filter = filtered_df['certification'] == 'PG-13'
rated_pg_filter = filtered_df['certification'] == 'PG'
rated_g_filter = filtered_df['certification'] == 'G'
```

```python
rated_r = filtered_df.loc[rated_r_filter]
rated_r['revenue'].mean().round(2)
```

```python
rated_pg = filtered_df.loc[rated_pg_filter]
rated_pg['revenue'].mean().round(2)
```

```python
rated_g = filtered_df.loc[rated_g_filter]
rated_g['revenue'].mean().round(2)
```

### Exploratory Visual

```python
fig, axes = plt.subplots(1,2, figsize=(8,4))
fig.suptitle('Budget and Revenue by Movie Rating')

sns.barplot(ax = axes[0], x = filtered_df['certification'],
            y = filtered_df['budget'])
axes[0].set_xlabel('Movie Rating', fontsize=(12))
sns.barplot(ax=axes[1], x = filtered_df['certification'],
           y = filtered_df['revenue'])
axes[1].set_xlabel('Movie Rating', fontsize=(12));
```
